## Convert string columns to numbers

- nominal --> One hot encoding (dummy variables)


- ordinal --> label encoding

```python
#label encoding
label_map={'BusinessTravel':{'Travel_Rarely':1,
            'Travel_Frequently':2,'Non-Travel':0}}
df.replace(label_map,inplace=True)
df['BusinessTravel'].unique()
```

```python
df.dtypes
```

```python
#OHE
df_ohe = pd.get_dummies(df)
df_ohe.shape,df.shape
```

```python
df_ohe.columns
```

```python
df_ohe.dtypes
```

# Fill missing values

```python
df_ohe.fillna(df_ohe.median(),inplace=True)
```

```python
df_ohe.isna().sum()
```

```python
cont_columns = df_ohe.columns[df_ohe.nunique() > 15]
cont_columns
```

```python
df_ohe.loc[:,cont_columns] = (df_ohe.loc[:,cont_columns] - df_ohe.loc[:,cont_columns].mean()) / df_ohe.loc[:,cont_columns].std()
```

```python
df_ohe.loc[:,cont_columns].mean()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df_ohe.loc[:,cont_columns] = sc.fit_transform(df_ohe.loc[:,cont_columns])
df_ohe.loc[:,cont_columns].mean()
```
