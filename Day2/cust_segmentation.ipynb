# Customer segmentation on E commerce data

https://www.kaggle.com/datasets/carrie1/ecommerce-data

```python
import pandas as pd
df = pd.read_csv("/content/customer_data.csv",
                 encoding="iso-8859-1")
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Description'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.columns
```

```python
df['Price'] = df['Quantity']*df['UnitPrice']
```

```python
df.columns
```

```python
df_1 = df.groupby(['InvoiceNo','InvoiceDate','CustomerID'])['Price'].sum().reset_index()
```

```python
df_1.head()
```

```python
df_1.shape
```

```python
df_1 = df_1.loc[df_1['Price'] > 0,:]
```

```python
df_1.shape
```

```python
# customer id , total no of invoices, total spent,
# std dev of spent in invoice
# days from last invoice
```

```python
df_2 = df_1.groupby(['CustomerID']).agg({'Price':'sum',
                                  'InvoiceNo':'count',
                                  'InvoiceDate':'max'}).reset_index()
```
