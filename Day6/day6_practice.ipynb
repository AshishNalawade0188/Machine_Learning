```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/ML_25/Topic6/devnagari_db_csv.zip")
```

```python
df['character'].unique()
```

```python
df1 = df.loc[df['character'].isin(['character_23_ba', 'character_24_bha',
 'character_25_ma', 'character_26_yaw',
 'character_29_waw','character_17_tha']), :]
df1.shape
```

```python
df1['character'].unique()
```

```python
X = df1.drop(columns = ['character'])
Y = df1['character']
X.shape,Y.shape
```

```python
label_map={'character_17_tha':0,
    'character_23_ba':1,
           'character_24_bha':2,
            'character_25_ma':3,
           'character_26_yaw':4,
            'character_29_waw':5,}
Y.replace(label_map,inplace=True)
Y.unique()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,
                                                 test_size=0.3,
                                                 random_state=7,
                                                 stratify=Y)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
# logistic regression ::
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=7)
lr.fit(X_train,Y_train)
Y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200,
                            oob_score=True,
                            max_samples=0.8,
                            max_features=1000,
                            random_state=7)
rf.fit(X_train, Y_train)
Y_pred = rf.predict(X_test)
print(classification_report(Y_test,Y_pred))
```

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier(n_estimators=50,
                    learning_rate=0.5,
                    max_depth=5,
                    subsample=0.8,
                       colsample_bytree=0.8,
                       random_state=7 )
xgb.fit(X_train,Y_train)
```

```python
Y_pred = xgb.predict(X_test)
print(classification_report(Y_test,Y_pred))
```

# PARAMETER TUNING

```python
df = pd.read_csv("/content/cleaned_76hd.csv")
df.shape
```

```python
X = df.drop(columns=['V58'])
Y = df['V58']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,
                                                 test_size=0.3,
                                                 random_state=7,
                                                 stratify=Y)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
pg = {
    'C':[0.1,0.5,1.0],
    'gamma':[0.5,1,5]
}
gscv = GridSearchCV(SVC(kernel ='rbf', random_state=7),
                    param_grid=pg, cv=2,
                    scoring='average_precision',
                    verbose=2)
gscv.fit(X_train,Y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
svm = SVC(C=1, gamma=0.5,kernel='rbf',random_state=7)
svm.fit(X_train,Y_train)
Y_pred = svm.predict(X_test)
print(classification_report(Y_test,Y_pred))
```
