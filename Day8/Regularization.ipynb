```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error, mean_absolute_error,mean_squared_error, r2_score
```

```python
df = pd.read_csv("/content/medical expenses.csv").dropna()
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.nunique().sum().sum()
```

```python
df.isna().sum()
```

```python
cont_col = ["age","bmi"]
```

```python
df_ohe = pd.get_dummies(df)
df_ohe.shape
```

```python
X = df_ohe.drop("expenses",axis=1)
Y = df_ohe["expenses"]
X.shape, Y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=7)
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train[cont_col] = sc.fit_transform(X_train[cont_col])
X_test[cont_col] = sc.transform(X_test[cont_col])
```

```python
X_train[cont_col].mean()
```

```python
lr = LinearRegression()
lr.fit(X_train,Y_train)
Y_pred = lr.predict(X_test)
```

```python
print("root mean squared error:",root_mean_squared_error(Y_test,Y_pred))
print("mean absolute error:",mean_absolute_error(Y_test,Y_pred))
print("r2_score :",r2_score(Y_test,Y_pred))
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# ploynomial regression
# 1.add ploy features in tha data
# 2.apply linear regression
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=2)
X_train_poly = pf.fit_transform(X_train)
X_test_poly = pf.transform(X_test)

X_train_poly.shape, X_test_poly.shape
```

```python
lr = LinearRegression()
lr.fit(X_train_poly,Y_train)
Y_pred = lr.predict(X_test_poly)
print("root mean squared error:",root_mean_squared_error(Y_test,Y_pred))
print("mean absolute error:",mean_absolute_error(Y_test,Y_pred))
print("r2_score :",r2_score(Y_test,Y_pred))
```

### Elastic Net Regression

```python
from sklearn.linear_model import ElasticNet

# Initialize and fit the Elastic Net model
en = ElasticNet(alpha=0.5, l1_ratio=0.5, random_state=7)
en.fit(X_train_poly3, Y_train)

Y_pred_en = en.predict(X_test_poly3)

# Evaluate the Elastic Net model
print("Elastic Net Root Mean Squared Error:", root_mean_squared_error(Y_test, Y_pred_en))
print("Elastic Net Mean Absolute Error:", mean_absolute_error(Y_test, Y_pred_en))
print("Elastic Net R2 Score:", r2_score(Y_test, Y_pred_en))
```
